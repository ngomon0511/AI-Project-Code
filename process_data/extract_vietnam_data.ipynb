Data link: https://ember-energy.org/data/monthly-electricity-data/

```python
import pandas as pd
```

```python
df = pd.read_csv("monthly_full_release_long_format.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df_vietnam = df[df['Area'] == 'Viet Nam']
```

```python
df_vietnam.head()
```

```python
# Take "Demand" on Subcategory
df_vietnam = df_vietnam[df_vietnam['Subcategory'] == 'Demand']
df_vietnam.rename(columns={'Value': 'Demand'}, inplace=True)

# Choose column Date and Demand
df_vietnam = df_vietnam[['Date', 'Demand']]

df_vietnam.head()
```

```python
# Save to csv
df_vietnam.to_csv('vietnam_electricity_demand.csv', index=False)
```
